## Перцептрон (один нейрон)

> Фундаментальный строительный блок нейросети.

<div class="d-flex align-items-center justify-content-center gap-10px">

![perceptron_ru.svg](imgs/perceptron_ru.svg)

<div>
<span>
$$
\overbrace{\hat{y}}^{\text{Выход}} = \overbrace{g\left(\underbrace{w_0}_{\text{Смещение}} + \sum_{i=1}^m \underbrace{x_i}_{\text{Вход}} \underbrace{w_i}_{\text{Вес}}\right)}^{\text{Нелинейная функция активации}}
$$
</span>

<hr />

<span>
$$
\hat{y}=g\left(w_0+\boldsymbol{X}^T \boldsymbol{W}\right)
$$

$$
\text{где:} \quad \boldsymbol{X}=\left[\begin{array}{c}x_1 \\ \vdots \\ x_m\end{array}\right] \quad \text{и} \quad \boldsymbol{W}=\left[\begin{array}{c}w_1 \\ \vdots \\ w_m\end{array}\right]
$$
</span>
</div>
</div>

### Однослойная нейросеть

![single_layer_nn_ru.svg](imgs/single_layer_nn_ru.svg)

### Пример однослойной нейросети

````python
@dataclass
class SLNN(nn.Module):
    input_size: int # Входной слой
    hidden_size: int # Скрытый слой
    output_size: int # Выходной слой

    # Вызывается сразу после __init__ при использовании dataclass
    def __post_init__(self):
        super(SLNN, self).__init__()

        # Определение скрытого слоя
        self.hidden = nn.Linear(self.input_size, self.hidden_size)
        # Определение выходного слоя
        self.output = nn.Linear(self.hidden_size, self.output_size)

        # Определение функции активации для скрытого слоя
        self.g = nn.ReLU()

    # Прямой проход данных через нейросеть
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        # Проход через скрытый слой с функцией активации
        x = self.g(self.hidden(x))
        # Проход через выходной слой
        x = self.output(x)

        return x

# Проверка наличия GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Создание экземпляра модели
model = SLNN(input_size = 10, hidden_size = 20, output_size = 7).to(device)

# Генерация случайного тензора
input = torch.randn(2, model.input_size).to(device)

# Проход через модель
output = model(input)

# Вывод результирующего тензора после прохождения через модель
print("Результирующий тензор после прохождения через модель:")
print(output)

# Вывод размера результирующего тензора
print("Размер результирующего тензора:")
print(output.size())
````

### Глубокая нейросеть

![deep_nn_ru.svg](imgs/deep_nn_ru.svg)

### Пример глубокой нейросети

````python
from dataclasses import dataclass
import torch
import torch.nn as nn

@dataclass
class DNN(nn.Module):
    input_size: int # Входной слой
    hidden_sizes: list[int] # Список размеров скрытых слоев
    output_size: int # Выходной слой

    def __post_init__(self):
        super(DNN, self).__init__()

        # Создание списка слоев
        layers = []
        in_features = self.input_size

        # Добавление скрытых слоев с функцией активации ReLU
        for hidden_size in self.hidden_sizes:
            layers.append(nn.Linear(in_features, hidden_size))
            layers.append(nn.ReLU())
            in_features = hidden_size

        # Добавление выходного слоя (без функции активации)
        layers.append(nn.Linear(in_features, self.output_size))

        # Объединение слоев в Sequential
        self.network = nn.Sequential(*layers)

    # Прямой проход через нейросеть
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.network(x)

# Проверка наличия GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Создание экземпляра модели с 3 скрытыми слоями
model = DNN(
    input_size = 10,
    hidden_sizes = [50, 30, 20], # Скрытые слои с 50, 30 и 20 нейронами
    output_size = 7
).to(device)

# Генерация случайного тензора
input = torch.randn(3, model.input_size).to(device)

# Проход через модель
output = model(input)

# Вывод результирующего тензора после прохождения через модель
print("Результирующий тензор после прохождения через глубокую модель:")
print(output)

# Вывод размера результирующего тензора
print("Размер результирующего тензора:")
print(output.size())
````

### Пример сверточной нейросети (CNN)

#### Когда применять CNN

1. **Анализ изображений**
   - Классификация изображений (например, распознавание объектов: кошка/собака)
   - Сегментация изображений (разделение на области, например, выделение дорог на спутниковых снимках)
   - Обнаружение объектов (определение местоположения объектов на изображении, например, YOLO)
2. **Видеоанализ**
   - Классификация видео (например, определение действия на видео)
   - Отслеживание объектов (например, движение автомобиля в кадре)
3. **Обработка текстов**
   - Эмбеддинги текста (с использованием сверточных фильтров для выделения локальных шаблонов)
   - Классификация текстов (например, анализ тональности)
4. **Обработка временных рядов**
   - Прогнозирование (например, временной анализ с учетом локальных шаблонов)
   - Аномалия в данных (обнаружение выбросов)
5. **Аудиообработка**
   - Классификация аудиосигналов (например, распознавание команд голосового ассистента)
   - Спектрограммы (применение CNN для анализа визуальных представлений звука)

#### Преимущества CNN

1. **Эффективность в обработке изображений**
   - Улавливают пространственные зависимости, такие как формы, границы и текстуры
   - Используют сверточные фильтры, которые `видят` локальные шаблоны
2. **Меньше параметров**
   - Свертки используют весовые коэффициенты повторно (shared weights), что снижает количество параметров
   - Например, в отличие от полносвязных слоев, где каждый нейрон связан с каждым входом, сверточные сети обучают только фильтры
3. **Инвариантность к сдвигам**
   - MaxPooling и свертки позволяют обрабатывать объекты на изображении, даже если они слегка сдвинуты или повернуты
4. **Универсальность**
   - Применяются к различным типам данных: изображения, аудио, текст, временные ряды

#### Недостатки CNN

1. **Большая вычислительная сложность**
   - Сверточные операции требуют значительных вычислительных ресурсов, особенно на высоких разрешениях (например, 4K изображения)
   - Потребность в GPU для ускорения обработки
2. **Требования к данным**
   - Требуется большое количество данных для обучения, чтобы избежать переобучения
   - Может быть трудно обучить сеть на небольших наборах данных
3. **Чувствительность к данным**
   - Плохо работают с шумными и плохо размеченными данными
   - Легко переобучаются, если данные однотипные или узконаправленные
4. **Сложность интерпретации**
   - Трудно понять, какие признаки были выделены на конкретных слоях
   - Требуются методы визуализации, чтобы интерпретировать работу сети

````python
@dataclass
class CNN(nn.Module):
    input_channels: int # Количество каналов во входных данных (например, 3 для RGB-изображений)
    num_classes: int # Количество классов для классификации

    def __post_init__(self):
        super(CNN, self).__init__()

        # Сверточные слои и пулинг
        # Слой 1: Сверточный слой с 16 фильтрами, ядром 3x3, шагом 1 и паддингом 1
        # Это позволяет сохранить исходные размеры изображения (28x28)
        # ReLU — функция активации, которая добавляет нелинейность
        # MaxPool2d уменьшает размерность входа в 2 раза
        self.conv_layers = nn.Sequential(
            nn.Conv2d(self.input_channels, 16, kernel_size = 3, stride = 1, padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2),

            # Слой 2: Еще один сверточный слой с 32 фильтрами, аналогичный первому
            # Это позволяет модели извлекать более сложные признаки из изображения
            nn.Conv2d(16, 32, kernel_size = 3, stride = 1, padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2)
        )

        # Полносвязный слой: принимает выход из сверточных слоев и преобразует его
        # в количество классов. В данном случае используется размерность 32*7*7,
        # так как после двух MaxPool размерность изображения уменьшается с 28x28 до 7x7
        self.fc = nn.Linear(32 * 7 * 7, self.num_classes)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        # Прямой проход через сверточные слои
        x = self.conv_layers(x)

        # Изменение размерности: Flatten выход из сверточных слоев
        # для передачи в полносвязный слой. x.view преобразует тензор
        x = x.view(x.size(0), -1)

        # Прямой проход через полносвязный слой
        x = self.fc(x)

        return x

# Проверка наличия GPU для ускорения вычислений
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Создание модели для работы с изображениями размером 28x28 с 3 каналами
model = CNN(input_channels = 3, num_classes = 10).to(device)

# Генерация случайных входных данных, имитирующих мини-батч изображений
# Тензор имеет размерность (8, 3, 28, 28): 8 изображений, 3 канала, размер 28x28
input = torch.randn(8, 3, 28, 28).to(device)

# Прямой проход данных через модель
output = model(input)

# Вывод результирующего тензора, содержащего вероятности принадлежности к каждому классу
print("Результирующий тензор после прохождения через CNN:")
print(output)

# Размер результирующего тензора будет (8, 10): 8 объектов, 10 классов
print("Размер результирующего тензора:")
print(output.size())
````

### Пример рекуррентной нейросети (RNN)

#### Когда применять RNN

1. **Анализ последовательностей текста**
   - Генерация текста (например, чат-боты, автозаполнение текста)
   - Анализ тональности (определение эмоциональной окраски текста)
   - Перевод текста (машинный перевод с одного языка на другой)
2. **Обработка временных рядов**
   - Прогнозирование (например, предсказание цен на акции)
   - Обнаружение аномалий (например, в данных о температуре или финансовых транзакциях)
3. **Аудиообработка**
   - Распознавание речи (например, преобразование аудио в текст)
   - Классификация звуков (например, определение типа звука — шаги, аплодисменты и т.д.)
4. **Видеоанализ**
   - Классификация действий (например, определение активности на видео)
   - Анализ движений (например, предсказание следующего кадра в видео)
5. **Другие последовательности**
   - Геномика (анализ последовательностей ДНК)
   - Лог-файлы (анализ системных событий или пользовательских данных)

#### Преимущества RNN

1. **Учет временных зависимостей**
   - Модели учитывают порядок данных, что особенно важно для последовательностей, таких как текст или временные ряды
2. **Обработка переменной длины данных**
   - RNN способны работать с входными последовательностями разной длины без необходимости предварительного их приведения к одному размеру
3. **Память прошлых состояний**
   - Учитывают контекст предыдущих данных через скрытые состояния, что полезно для задач, требующих исторической информации
4. **Универсальность**
   - Применимы к разнообразным типам данных, включая текст, аудио, видео, временные ряды

#### Недостатки RNN

1. **Проблема градиентов**
   - Трудности с обучением, связанные с затуханием или взрывом градиентов при обработке длинных последовательностей
   - Это делает обучение стандартных RNN менее эффективным для задач с долгосрочными зависимостями
2. **Высокие вычислительные затраты**
   - Из-за последовательной обработки (step-by-step) обучение и инференс медленнее по сравнению с другими архитектурами, такими как трансформеры
3. **Проблемы с долгосрочной памятью**
   - Сложности с запоминанием событий, происходивших в далеком прошлом последовательности
4. **Требовательность к данным**
   - Требуется тщательная обработка данных и их представление для обучения

#### Преимущества специализированных версий RNN

1. **LSTM**
   - Решает проблему градиентов
   - Применяется для задач, где важны долгосрочные зависимости
2. **GRU**
   - Упрощенная версия LSTM с меньшим количеством параметров
   - Быстрее обучается и лучше работает на небольших наборах данных

#### Аналоги и альтернативы RNN

1. **CNN для временных рядов**
   - Используются для выделения локальных шаблонов, но не сохраняют глобальный контекст
2. **Transformers**
   - Основаны на механизме внимания (attention), который хорошо работает с длинными зависимостями
   - Например, модели вроде BERT, GPT показывают хорошие результаты в обработке текста
3. **xLSTM**
   - Расширение LSTM для работы с еще более длинными последовательностями, улучшая производительность в задачах с долгосрочной памятью
4. **Mamba**
   - Архитектура, комбинирующая внимание и память для повышения точности в задачах с длинными последовательностями
5. **WaveNet**
   - Используется для обработки временных данных, таких как аудио, и генерирует высококачественные результаты благодаря глубоким сверточным слоям

````python
@dataclass
class RNN(nn.Module):
    input_size: int # Размерность входного вектора, например, количество признаков в каждом временном шаге
    hidden_size: int # Размер скрытого состояния (количество нейронов в скрытом слое)
    output_size: int # Размерность выходного вектора (количество классов или значений на выходе)
    num_layers: int = 1 # Количество слоев в RNN, по умолчанию 1

    def __post_init__(self):
        super(RNN, self).__init__()

        # Рекуррентный слой (Simple RNN)
        # batch_first=True означает, что входные данные имеют формат (batch_size, sequence_length, input_size)
        # Это упрощает работу с последовательными данными
        self.rnn = nn.RNN(
            input_size = self.input_size,
            hidden_size = self.hidden_size,
            num_layers = self.num_layers,
            batch_first = True
        )

        # Полносвязный слой: преобразует скрытое состояние RNN в выходное пространство
        # Например, для классификации на 5 классов размерность будет self.output_size=5
        self.fc = nn.Linear(self.hidden_size, self.output_size)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        # Прямой проход через рекуррентный слой
        # out содержит выходные состояния для каждого временного шага
        # _ содержит финальное скрытое состояние (может быть полезно для задач на уровне последовательностей)
        out, _ = self.rnn(x)

        # Преобразование выхода последнего временного шага через полносвязный слой
        # out[:, -1, :] означает, что берется выход на последнем временном шаге для каждого объекта в батче
        out = self.fc(out[:, -1, :])

        return out

# Проверка наличия GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Создание модели RNN
# Входные данные имеют размерность 10, скрытое состояние — 20, выход — 5 классов
# Используется 2 слоя RNN для большей глубины
model = RNN(input_size = 10, hidden_size = 20, output_size = 5, num_layers = 2).to(device)

# Генерация случайного тензора, имитирующего последовательности
# Размер тензора: (batch_size=8, sequence_length=15, input_size=10)
input = torch.randn(8, 15, 10).to(device)

# Прямой проход данных через модель
output = model(input)

# Результирующий тензор содержит выходы для каждого объекта
# Размер тензора: (8, 5), где 8 — количество объектов, 5 — количество классов
print("Результирующий тензор после прохождения через RNN:")
print(output)

# Вывод размерности результирующего тензора
print("Размер результирующего тензора:")
print(output.size())
````

````python
# Токенизация текста
tokenizer = get_tokenizer("basic_english")
text = "PyTorch is an open-source machine learning framework"
tokens = tokenizer(text)

print("Токены:", tokens)

# Создание словаря с порядком индексов
def create_vocab_from_tokens(tokens, specials=["<unk>"]):
    """Создание словаря, следуя порядку токенов в тексте"""

    unique_tokens = list(dict.fromkeys(tokens)) # Сохранение порядка и удаление дубликатов
    ordered_dict = OrderedDict((token, 1) for token in unique_tokens) # Частота 1 для всех токенов
    vocab_instance = vocab(ordered_dict) # Создание объекта vocab без специальных токенов
    for special in specials: # Добавление специальные токены вручную
        vocab_instance.insert_token(special, 0) # Вставка в начало словаря
    vocab_instance.set_default_index(vocab_instance[specials[0]]) # Установка индекса по умолчанию
    return vocab_instance

# Создание словаря
vocab_instance = create_vocab_from_tokens(tokens)

# Индексы токенов
indices = [vocab_instance[token] for token in tokens]
print("Индексы токенов:", indices)

# Словарь
print("Словарь:", vocab_instance.get_stoi())
````

### Преобразование аудио в спектрограмму с Torchaudio

````python
# Загрузка аудиофайла
audio_path = "wavs/1624615136581_-2425.wav"
waveform, sample_rate = torchaudio.load(audio_path)

# Преобразование в спектрограмму
# Параметры:
# n_fft: количество точек для преобразования Фурье (определяет частотное разрешение)
# win_length: длина окна анализа (в отсчетах, влияет на временное разрешение)
# hop_length: шаг смещения окна (в отсчетах, определяет перекрытие окон)
spectrogram_transform = Spectrogram(n_fft = 400, win_length = 400, hop_length = 160)
spectrogram = spectrogram_transform(waveform)

# Функция для визуализации спектрограммы
def plot_spectrogram(spectrogram, sample_rate, title = "Спектрограмма", cmap = "viridis"):
    """Визуализация спектрограммы с логарифмической шкалой мощности"""

    plt.figure(figsize = (10, 6))
    for channel in range(spectrogram.shape[0]): # Для каждого канала (моно/стерео)
        plt.subplot(1, spectrogram.shape[0], channel + 1)
        plt.imshow(
            spectrogram[channel].log2().numpy(), # Преобразование в логарифмическую шкалу
            aspect = "auto", # Масштабирование по осям
            origin = "lower", # Начало координат внизу
            cmap = cmap
        )
        plt.title(f"{title} - Канал {channel + 1}")
        plt.xlabel("Время (кадры)")
        plt.ylabel("Частота (бины)")
        plt.colorbar(label="Логарифм мощности")

    plt.tight_layout()
    plt.show()

# Вывод формы спектрограммы и визуализация
print("Форма спектрограммы:", spectrogram.shape) # [каналы, частота, время]
plot_spectrogram(spectrogram, sample_rate)
````

### Объединение данных в мультимодальный батч (версия 2)

1. **Поддержка метаинформации.** Добавление возможности работы с дополнительной метаинформацией (например, теги, категории или временные метки), которая может быть полезна для дальнейшего анализа или обучения.
2. **Аугментация данных.** Добавление аугментации для изображений, аудио и текста, чтобы улучшить устойчивость модели к разнообразию данных
3. **Сохранение и кеширование обработанных данных.** Кеширование обработанных данных, чтобы избежать повторной обработки больших файлов (например, аудио и изображений)
4. **Динамическая длина батча.** Использование подхода `динамического формирования батча` для текстов и аудио, где элементы объединяются по длине, чтобы уменьшить объем padding (отступы)

````python
@dataclass
class MultimodalDataset(Dataset):
    """Расширенный класс для мультимодальных данных с метаинформацией, аугментацией и поддержкой кеширования

    Атрибуты:
        image_paths (List[str]): Пути к файлам изображений
        audio_paths (List[str]): Пути к аудиофайлам
        texts (List[str]): Список текстовых данных (строки)
        metadata (Optional[List[Dict]]): Дополнительная метаинформация (опционально)
        transform (transforms.Compose): Трансформация для обработки изображений
        resample_rate (int): Частота дискретизации для аудио (по умолчанию 16 kHz)
        cache_dir (Optional[str]): Директория для кеша обработанных данных
    """

    image_paths: List[str]
    audio_paths: List[str]
    texts: List[str]
    metadata: Optional[List[Dict]] = None
    transform: transforms.Compose = transforms.Compose([transforms.Resize((128, 128)), transforms.ToTensor()])
    resample_rate: int = 16000
    cache_dir: Optional[str] = None

    def __post_init__(self):
        # Проверка, существует ли директория для кеша
        if self.cache_dir and not os.path.exists(self.cache_dir):
            os.makedirs(self.cache_dir)

    def __len__(self) -> int:
        """Возвращает количество элементов в наборе данных"""

        return len(self.texts)

    def _augment_audio(self, audio_tensor: torch.Tensor) -> torch.Tensor:
        """Аугментация для аудио: добавление шума"""

        noise = torch.randn_like(audio_tensor) * 0.005

        return audio_tensor + noise

    def _augment_text(self, text: str) -> str:
        """Аугментация текста: перестановка слов"""

        words = text.split()
        random.shuffle(words)

        return " ".join(words)

    def _load_or_cache(self, idx: int, cache_key: str, loader_func):
        """Загрузка данных из кеша или выполнение обработки"""

        if self.cache_dir:
            cache_path = os.path.join(self.cache_dir, f"{cache_key}_{idx}.pt")
            if os.path.exists(cache_path):
                return torch.load(cache_path)

        data = loader_func()

        if self.cache_dir:
            torch.save(data, cache_path)

        return data

    def __getitem__(self, idx: int):
        """Загрузка и обработка данных по индексу"""

        # Обработка изображения
        image_tensor = self._load_or_cache(idx, "image", lambda: self.transform(Image.open(self.image_paths[idx])))

        # Обработка аудио
        waveform, sample_rate = torchaudio.load(self.audio_paths[idx])
        resampler = Resample(orig_freq = sample_rate, new_freq = self.resample_rate)
        audio_tensor = self._load_or_cache(
            idx, "audio", lambda: resampler(waveform).squeeze(0)
        )
        audio_tensor = self._augment_audio(audio_tensor)

        # Обработка текста
        tokenizer = get_tokenizer("basic_english")
        tokens = tokenizer(self._augment_text(self.texts[idx]))
        vocab = build_vocab_from_iterator([tokens], specials = ["<unk>"])
        vocab.set_default_index(vocab["<unk>"])
        text_tensor = torch.tensor([vocab[token] for token in tokens])

        # Метаинформация (если есть)
        meta = self.metadata[idx] if self.metadata else {}

        return image_tensor, audio_tensor, text_tensor, meta


def collate_fn(batch):
    """Функция для обработки батчей мультимодальных данных с поддержкой паддинга аудио и текстов
    """
    images, audios, texts, metas = zip(*batch)

    # Паддинг аудио
    max_audio_length = max([audio.size(0) for audio in audios])
    audios = [F.pad(audio, (0, max_audio_length - audio.size(0))) for audio in audios]

    # Паддинг текстов
    texts = pad_sequence(texts, batch_first = True, padding_value = 0)

    return torch.stack(images), torch.stack(audios), texts, metas


# Пример использования
image_paths = ["imgs/example.jpg"] * 4
audio_paths = ["wavs/1624615136581_-2425.wav"] * 4
texts = ["Hello world", "PyTorch", "Deep Learning", "Multimodal Corpora"]
metadata = [{"category": "A"}, {"category": "B"}, {"category": "C"}, {"category": "D"}]

dataset = MultimodalDataset(
    image_paths = image_paths,
    audio_paths = audio_paths,
    texts = texts,
    metadata = metadata,
    cache_dir = "./cache"
)

data_loader = DataLoader(dataset, batch_size = 2, shuffle = True, collate_fn = collate_fn)

for batch in data_loader:
    images, audios, texts, metas = batch
    print("Батч изображений:", images.shape)
    print("Батч аудио:", audios.shape)
    print("Батч текстов:", texts.shape)
    print("Метаинформация:", metas)
````

#### Cross-Attention для мультимодальной обработки

> Cross-Attention позволяет моделям `фокусироваться` на важных признаках из разных модальностей. Например, аудио может выделить важные слова в тексте, а изображение может помочь уточнить контекст

````python
# Класс для хранения признаков
@dataclass
class ModalFeatures:
    image: torch.Tensor
    audio: torch.Tensor
    text: torch.Tensor

# Класс для модели с Cross-Attention
@dataclass
class CrossAttentionModel:
    input_dim: int
    output_dim: int
    attn: nn.Module = field(init = False)
    fc: nn.Module = field(init = False)

    def __post_init__(self):
        """Инициализация слоев после создания объекта модели"""

        # Инициализируем слои внутри __post_init__
        self.attn = nn.MultiheadAttention(embed_dim = self.input_dim, num_heads = 4)
        self.fc = nn.Linear(self.input_dim, self.output_dim)

    def forward(self, features: ModalFeatures):
        """Применение межмодального внимания и объедение признаков всех модальностей

        :param features: Экземпляр ModalFeatures, содержащий признаки для изображений, аудио и текста
        :return: Объединенные признаки после применения внимания и линейного слоя
        """

        # Конкатенация признаков всех модальностей
        concatenated_features = torch.cat([features.image, features.audio, features.text], dim = 1)

        # Добавление размерности для пакета
        concatenated_features = concatenated_features.unsqueeze(0) # (1, batch_size, input_dim)

        # Применение MultiheadAttention для объединения модальностей
        attn_output, _ = self.attn(concatenated_features, concatenated_features, concatenated_features)

        # Удаление размерности пакета
        attn_output = attn_output.squeeze(0) # (batch_size, input_dim)

        # Применение линейного слоя для получения окончательных признаков
        return self.fc(attn_output)

# Пример использования:
# Создание признаков для разных модальностей
image_features = torch.randn(2, 512) # Признаки для изображений
audio_features = torch.randn(2, 128) # Признаки для аудио
text_features = torch.randn(2, 256) # Признаки для текста

# Упаковка признаков в dataclass
features = ModalFeatures(image = image_features, audio = audio_features, text = text_features)

# Инициализация модели с объединенной размерностью всех признаков
model = CrossAttentionModel(input_dim = 512 + 128 + 256, output_dim = 64)

# Получение объединенных признаков после применения CrossAttention
fused_features = model.forward(features)

print(f"Результат: {fused_features.shape}")
````

#### Использование Modal Fusion для эффективного смешивания модальностей

> Метод комбинирования различных типов данных в одну единую структуру признаков, используя параллельные потоки обработки для каждой модальности

````python
@dataclass
class ModalFusion(nn.Module):
    """
    Модель для объединения признаков различных модальностей:
    изображений, аудио и текста.
    Признаки каждой модальности сначала преобразуются в общее пространство (fusion_dim),
    затем объединяются через конкатенацию.
    """
    image_dim: int  # Размерность признаков изображения
    audio_dim: int  # Размерность признаков аудио
    text_dim: int   # Размерность признаков текста
    fusion_dim: int # Размерность общего пространства для каждой модальности

    # Поля, которые будут инициализированы в методе __post_init__
    image_fc: nn.Module = field(init=False)
    audio_fc: nn.Module = field(init=False)
    text_fc: nn.Module = field(init=False)

    def __post_init__(self):
        """
        Инициализация линейных слоев для каждой модальности после создания объекта.
        Линейные преобразования уменьшают размерность признаков
        каждой модальности до fusion_dim.
        """
        super().__init__()
        self.image_fc = nn.Linear(self.image_dim, self.fusion_dim)
        self.audio_fc = nn.Linear(self.audio_dim, self.fusion_dim)
        self.text_fc = nn.Linear(self.text_dim, self.fusion_dim)

    def forward(self, image_features: torch.Tensor, audio_features: torch.Tensor, text_features: torch.Tensor) -> torch.Tensor:
        """
        Прямой проход через модель.

        :param image_features: Признаки изображения (batch_size, image_dim)
        :param audio_features: Признаки аудио (batch_size, audio_dim)
        :param text_features: Признаки текста (batch_size, text_dim)
        :return: Объединенные признаки (batch_size, fusion_dim * 3)
        """
        # Преобразуем каждую модальность в общее пространство
        image_output = self.image_fc(image_features)  # (batch_size, fusion_dim)
        audio_output = self.audio_fc(audio_features)  # (batch_size, fusion_dim)
        text_output = self.text_fc(text_features)    # (batch_size, fusion_dim)

        # Конкатенируем все модальности
        fused_output = torch.cat([image_output, audio_output, text_output], dim=1)  # (batch_size, fusion_dim * 3)

        return fused_output

# Пример использования
# Размерности входных признаков для изображений, аудио и текста
image_dim = 512
audio_dim = 128
text_dim = 256
fusion_dim = 64

# Инициализируем модель
model = ModalFusion(image_dim=image_dim, audio_dim=audio_dim, text_dim=text_dim, fusion_dim=fusion_dim)

# Пример входных данных (batch_size = 2)
image_features = torch.randn(2, image_dim)  # Признаки изображения
audio_features = torch.randn(2, audio_dim)  # Признаки аудио
text_features = torch.randn(2, text_dim)    # Признаки текста

# Получаем объединённые признаки
fused_features = model(image_features, audio_features, text_features)

# Вывод результатов
print(f"Размерность входных признаков изображения: {image_features.shape}")
print(f"Размерность входных признаков аудио: {audio_features.shape}")
print(f"Размерность входных признаков текста: {text_features.shape}")
print(f"Размерность объединённых признаков: {fused_features.shape}")
````

### Почему размер выходных признаков `torch.Size([2, 192])`?

#### Входные данные

Размерности входных данных для каждой модальности:
   - Изображения: $[\text{batch\_size}, \text{image\_dim}] = [2, 512]$
   - Аудио: $[\text{batch\_size}, \text{audio\_dim}] = [2, 128]$
   - Текст: $[\text{batch\_size}, \text{text\_dim}] = [2, 256]$
   
   `batch_size = 2` означает, что модель обрабатывает два примера одновременно

#### Преобразование через линейные слои

Каждый вектор признаков преобразуется в общее пространство размерности $\text{fusion\_dim} = 64$ с помощью линейных слоев:
   - Для **изображений**:
     $[2, 512] \xrightarrow{\text{image\_fc (Linear)}} [2, 64]$
   - Для **аудио**:
     $[2, 128] \xrightarrow{\text{audio\_fc (Linear)}} [2, 64]$
   - Для **текста**:
     $[2, 256] \xrightarrow{\text{text\_fc (Linear)}} [2, 64]$

#### Объединение признаков (конкатенация)

После преобразования каждый вектор имеет размерность $[2, 64]$. Эти векторы объединяются вдоль оси признаков $\text{dim}=1$ с помощью `torch.cat`:
   $
   \text{fused\_output} = \text{cat}([\text{image\_output}, \text{audio\_output}, \text{text\_output}], \text{dim}=1)
   $

Размерности объединяемых признаков:
   $
   [2, 64] + [2, 64] + [2, 64] = [2, 192]
   $

#### Итоговая размерность

После конкатенации объединённый вектор имеет размерность:
   $
   [\text{batch\_size}, \text{fusion\_dim} \times 3] = [2, 192]
   $
   
   где:
   - $\text{batch\_size} = 2$ — количество примеров в одном батче
   - $\text{fusion\_dim} = 64$ — размерность преобразованных признаков для каждой модальности
   - $3$ — количество модальностей (изображение, аудио, текст)

#### Итог

Размер выходных признаков равен `torch.Size([2, 192])` из-за конкатенации трех преобразованных векторов $[2, 64]$, соответствующих изображениям, аудио и тексту. Каждый из них имеет одинаковую размерность, заданную параметром $\text{fusion\_dim}$

### Объединение Модальностей с использованием MultiheadAttention

````python
@dataclass
class AttentionFusion(nn.Module):
    """Модуль AttentionFusion, объединяющий признаки из различных модальностей с использованием механизма MultiheadAttention
    
    Параметры:
    - image_dim (int): Размерность входных признаков изображения
    - audio_dim (int): Размерность входных признаков аудио
    - text_dim (int): Размерность входных признаков текста
    - fusion_dim (int): Общая размерность для преобразованных признаков
    """

    image_dim: int
    audio_dim: int
    text_dim: int
    fusion_dim: int

    def __post_init__(self):
        super().__init__()

        # Линейные преобразования для каждой модальности
        self.image_fc = nn.Linear(self.image_dim, self.fusion_dim)
        self.audio_fc = nn.Linear(self.audio_dim, self.fusion_dim)
        self.text_fc = nn.Linear(self.text_dim, self.fusion_dim)

        # Механизм MultiheadAttention
        self.attn = nn.MultiheadAttention(embed_dim = self.fusion_dim, num_heads = 1)

        # Финальное линейное преобразование (например, классификатор)
        self.fc = nn.Linear(self.fusion_dim, self.fusion_dim)

    def forward(self, image_features: torch.Tensor, audio_features: torch.Tensor, text_features: torch.Tensor) -> torch.Tensor:
        """Прямой проход через модель

        Аргументы:
        - image_features (torch.Tensor): Признаки изображения размерности (batch_size, image_dim)
        - audio_features (torch.Tensor): Признаки аудио размерности (batch_size, audio_dim)
        - text_features (torch.Tensor): Признаки текста размерности (batch_size, text_dim)

        Возвращает:
        - torch.Tensor: Признаки, объединенные с помощью внимания, размерности (batch_size, fusion_dim)
        """

        # Преобразование каждой модальности в общее пространство размерности fusion_dim
        image_output = self.image_fc(image_features).unsqueeze(0) # (1, batch_size, fusion_dim)
        audio_output = self.audio_fc(audio_features).unsqueeze(0) # (1, batch_size, fusion_dim)
        text_output = self.text_fc(text_features).unsqueeze(0) # (1, batch_size, fusion_dim)

        # Конкатенация всех модальностей вдоль временной оси
        combined_features = torch.cat([image_output, audio_output, text_output], dim = 0) # (3, batch_size, fusion_dim)

        # Применение механизма MultiheadAttention
        attn_output, _ = self.attn(combined_features, combined_features, combined_features)

        # Среднее по временной оси
        fused_features = attn_output.mean(dim = 0) # (batch_size, fusion_dim)

        # Финальное линейное преобразование
        return self.fc(fused_features)

# Пример использования
# Размерности входных данных
image_features = torch.randn(2, 512) # 2 примера, размерность 512
audio_features = torch.randn(2, 128) # 2 примера, размерность 128
text_features = torch.randn(2, 256) # 2 примера, размерность 256

# Создание модели AttentionFusion
model = AttentionFusion(image_dim = 512, audio_dim = 128, text_dim = 256, fusion_dim = 64)

# Получение объединённых признаков
fused_features = model(image_features, audio_features, text_features)

# Вывод результата
print(f"Размерность входных признаков изображения: {image_features.shape}")
print(f"Размерность входных признаков аудио: {audio_features.shape}")
print(f"Размерность входных признаков текста: {text_features.shape}")
print(f"Размерность объединённых признаков: {fused_features.shape}")
````
